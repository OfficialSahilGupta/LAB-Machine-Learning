```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

abs_df = df.set_index("Variable").abs()
principal_vars = abs_df.idxmax()
print("Principal component variables:\n", principal_vars)

fig, ax = plt.subplots(figsize=(8,5))
df.set_index("Variable").plot(kind="bar", ax=ax)
plt.title("PCA Loadings")
plt.ylabel("Loading Value")
plt.xlabel("Variables")
plt.xticks(rotation=45)
plt.legend(title="Principal Components")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

abs_df = df.set_index("Variable").abs()
principal_vars = abs_df.idxmax()

print("Principal component variables:\n")
for pc in principal_vars.index:
    print(f"{pc}: {principal_vars[pc]}")

fig, ax = plt.subplots(figsize=(9,6))
df.set_index("Variable").plot(kind="bar", ax=ax)
plt.title("PCA Loadings per Variable")
plt.ylabel("Loading Value")
plt.xlabel("Variables")
plt.xticks(rotation=45)
plt.legend(title="Principal Components")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
