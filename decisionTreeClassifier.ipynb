```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

np.random.seed(42)
n_samples = 200

ages = np.random.randint(20, 61, size=n_samples)
incomes = np.random.randint(30000, 100001, size=n_samples)

buy_house = []
for age in ages:
    if age > 30:
        buy_house.append(np.random.choice([1, 0], p=[0.7, 0.3]))
    else:
        buy_house.append(np.random.choice([1, 0], p=[0.3, 0.7]))

df_house = pd.DataFrame({'Age': ages, 'Income': incomes, 'BuyHouse': buy_house})

X_house = df_house[['Age', 'Income']]
y_house = df_house['BuyHouse']

clf_house = DecisionTreeClassifier(max_depth=3, random_state=42)
clf_house.fit(X_house, y_house)

plt.figure(figsize=(12, 8))
plot_tree(clf_house, feature_names=['Age', 'Income'], class_names=['No', 'Yes'], filled=True)
plt.title('Decision Tree - House Buying')
plt.show()

plt.clf()  # Clear the current figure

data_activity = {'Weather': ['Sunny', 'Sunny', 'Sunny', 'Rainy', 'Rainy', 'Rainy'],
                 'Activity': ['Go Out', 'Go Out', 'Go Out', 'Stay In', 'Stay In', 'Stay In']}

df_activity = pd.DataFrame(data_activity)
df_activity['Weather_Num'] = df_activity['Weather'].map({'Sunny': 1, 'Rainy': 0})

X_activity = df_activity[['Weather_Num']]
y_activity = df_activity['Activity']

clf_activity = DecisionTreeClassifier(random_state=42)
clf_activity.fit(X_activity, y_activity)

plt.figure(figsize=(6, 4))
plot_tree(clf_activity, feature_names=['Weather'], class_names=clf_activity.classes_, filled=True)
plt.title('Decision Tree - Activity Based on Weather')
plt.show()
```
